```python
import plumbum
```

```python
import pandas as pd
```

```python
dfs = []

for n in [20, 30, 50]:
    for m in [0, int(0.5 * n), n, int(1.5 * n), int(2 * n), int(3 * n)]:
        for method in [ 'dfs', 'gadag' ]:
            for connectivity in [1, 2, 3]:
                number = 2
                
                res_file = 'res.csv'
                cmd = plumbum.local['./scripts/run_tests.py'][
                    '-n', n
                ][
                    '-m', n * connectivity + m
                ][
                    '--cnt', 50
                ][
                    '--connectivity', connectivity
                ][
                    '--method', method
                ][
                    '--number', number
                ][
                    '--result', res_file
                ]
                
                print(f'Running: {cmd}')
                
                cmd.run()

                dfs.append(pd.read_csv(res_file))
                
df = pd.concat(dfs)

df.to_csv('all_res.csv', index=None)
```
